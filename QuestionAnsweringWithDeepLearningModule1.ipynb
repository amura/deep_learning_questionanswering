```python
import re
paragraphs = re.split("\n\n",text.decode('utf-8'))
```

```python
print(paragraphs)
```

```python
def toInclude(parag):  
  #return True
  return not parag.isspace()

def clean(txt):  
  txt =  re.sub(r"\n",r"",txt)
  #txt = re.sub(r"', '",r"",x)
  txt = re.sub(r"[^\w\s]", "", txt)
  return txt
```

```python
paragraphs = [clean(x) for x in paragraphs if toInclude(x) ]
#paragraphs = [re.sub(r"', '",r"",x) for x in paragraphs if toInclude(x)]
```

```python
paragraphs = pd.DataFrame(paragraphs, columns=['paragraphs'])
```
